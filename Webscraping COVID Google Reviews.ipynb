```python
from google_play_scraper import Sort, reviews_all
import pprint
pp = pprint.PrettyPrinter(indent=2)
import csv
import pandas as pd
```

```python
links = ['gov.adph.exposurenotifications', 'gov.azdhs.covidwatch.android', 'gov.co.cdphe.exposurenotifications','gov.ct.covid19.exposurenotifications', 'gov.de.covidtracker', 'gov.md.covid19.exposurenotifications', 'gov.michigan.MiCovidExposure', 'org.pathcheck.covidsafepathsBt.mn', 'gov.nv.dhhs.en', 'com.nj.gov.covidalert', 'gov.ny.health.proximity', 'gov.nc.dhhs.exposurenotification', 'com.proudcrowd.exposure', 'gov.pa.covidtracker', 'gov.vdh.exposurenotification', 'gov.wa.doh.exposurenotifications', 'com.proudcrowd.exposure', 'gov.md.covid19.exposurenotifications']
csvs = ['Alabama_Google_Reviews.csv','Arizona_Google_Reviews.csv','Colorado_Google_Reviews.csv','Connecticut_Google_Reviews.csv','Delaware_Google_Reviews.csv','Maryland_Google_Reviews.csv','Michigan_Google_Reviews.csv','Minnesota_Google_Reviews.csv','Nevada_Google_Reviews.csv','NewJersey_Google_Reviews.csv','NewYork_Google_Reviews.csv','NorthCarolina_Google_Reviews.csv','NorthDakota_Google_Reviews.csv','Pennsylvania_Google_Reviews.csv','Virginia_Google_Reviews.csv','Washington_Google_Reviews.csv','Wyoming_Google_Reviews.csv','WashingtonDC_Google_Reviews.csv']
```

```python
for i in range (len(links)):
    result = []
    result = reviews_all(
    links[i],
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=None # defaults to None(means all score)
)
    userID = []
    reviewId = []
    stars = []
    date = []
    review = []
    num = 1
    for j in range(len(result)):
        userID.append(num)
        reviewId.append(result[j]['reviewId'])
        stars.append(result[j]['score'])
        date.append(result[j]['at'])
        review.append(result[j]['content'])
        num += 1
        
    pairs = {'User ID': userID, 'URL': reviewId, 'Stars': stars, 'Date': date, 'Review': review}

    df = pd.DataFrame.from_dict(pairs)
    df.to_csv(csvs[i])
    
```
