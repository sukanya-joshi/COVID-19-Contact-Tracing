```python
from app_store_scraper import AppStore
from pprint import pprint
import pprint
pp = pprint.PrettyPrinter(indent=2)
import csv
import pandas as pd
```

```python
csvs = ['Alabama_Apple_Reviews.csv','Arizona_Apple_Reviews.csv','Delaware_Apple_Reviews.csv','Michigan_Apple_Reviews.csv','Nevada_Apple_Reviews.csv','NewJersey_Apple_Reviews.csv','NewYork_Apple_Reviews.csv','NorthCarolina_Apple_Reviews.csv','NorthDakota_Apple_Reviews.csv','Pennsylvania_Apple_Reviews.csv','Virginia_Apple_Reviews.csv','Wyoming_Apple_Reviews.csv']
appname = ["GuideSafe","Covid Watch Arizona","COVID Alert DE","MI COVID Alert","COVID Trace Nevada","COVID Alert NJ","COVID Alert NY","SlowCOVIDNC","Care19 Alert","COVID Alert Pennsylvania","COVIDWISE","Care19 Alert"]
appId = [1519514691,1521655110,1527783171,1527644912,1524365131,1529622525,1524123298,1526471580,1513945072,1527125511,1518059690,1513945072]
```

```python
for i in range (len(csvs)):
    result = AppStore(country="us", app_name = appname[i], app_id = appId[i])
    result.review(how_many = 100000)
    
    pp.pprint(result.reviews)
    
    userID = []
    username = []
    title = []
    stars = []
    date = []
    review = []
    num = 1

    for j in range(result.reviews_count):
            userID.append(num)
            username.append(result.reviews[j]['userName'])
            stars.append(result.reviews[j]['rating'])
            title.append(result.reviews[j]['title'])
            date.append(result.reviews[j]['date'])
            review.append(result.reviews[j]['review'])
            num += 1
        
    pairs = {'User ID': userID, 'Username': username, 'Title': title, 'Stars': stars, 'Date': date, 'Review': review}

    df = pd.DataFrame.from_dict(pairs)
    df.to_csv(csvs[i])
```
